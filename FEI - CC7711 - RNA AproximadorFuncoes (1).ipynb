```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

%matplotlib inline
```

```python
arquivo = np.load('teste1.npy')
x = arquivo[0]
y = arquivo[1]

plt.figure(figsize=(16,8))
plt.plot(x,y)
plt.show()
```

```python
from sklearn.neural_network import MLPRegressor

regr = MLPRegressor(hidden_layer_sizes=(2,2),
                    max_iter=10000,
                    solver='adam',
                    learning_rate = 'adaptive',
                    n_iter_no_change=1000)
regr = regr.fit(x,y)  

y_est = regr.predict(x)

#print da curva de aprendizagem
plt.figure(figsize=(16,8))
plt.plot(regr.loss_curve_)
plt.show()


#preditor
y_est = regr.predict(x)
plt.figure(figsize=(16,8))
plt.plot(x,y,x,y_est)
plt.show()
```

```python
arquivo = np.load('teste2.npy')
x = arquivo[0]
y = arquivo[1]

array1 = []
array2 = []
array3 = []
plt.figure(figsize=(16,8))
plt.plot(x,y)
plt.show()

from sklearn.neural_network import MLPRegressor

regr = MLPRegressor(hidden_layer_sizes=(50,5),
                    max_iter=10000,
                    solver='adam',
                    learning_rate = 'adaptive',
                    n_iter_no_change=1000)

regr3 = MLPRegressor(hidden_layer_sizes=(50,100),
                    max_iter=1000,
                    solver='adam',
                    learning_rate = 'adaptive',
                    n_iter_no_change=10000)

regr2 = MLPRegressor(hidden_layer_sizes=(4,4),
                    max_iter=1000,
                    solver='adam',
                    learning_rate = 'adaptive',
                    n_iter_no_change=10000)



savepoint = regr.fit(x,y)
savepoint2 = regr2.fit(x,y)
savepoint3 = regr3.fit(x,y)

array1.append(savepoint)
array2.append(savepoint2)
array3.append(savepoint3)

for count in range(9):
    regr = regr.fit(x,y)  
    regr2 = regr2.fit(x,y)
    regr3 = regr3.fit(x,y)

    array1.append(regr)
    array2.append(regr2)
    array3.append(regr3)
    
    if savepoint.loss < regr.loss:
        savepoint = regr
    if savepoint2.loss < regr2.loss:
        savepoint = regr2
    if savepoint3.loss < regr3.loss:
        savepoint = regr3

    

y_est = savepoint.predict(x)
y_est2 = savepoint.predict(x)
y_est3 = savepoint.predict(x)


#print da curva de aprendizagem
plt.figure(figsize=(16,8))
plt.plot(savepoint.loss_curve_)
plt.show()

#print da curva de aprendizagem
plt.figure(figsize=(16,8))
plt.plot(savepoint2.loss_curve_)
plt.show()

#print da curva de aprendizagem
plt.figure(figsize=(16,8))
plt.plot(savepoint3.loss_curve_)
plt.show()

#preditor
y_est = savepoint.predict(x)
plt.figure(figsize=(16,8))
plt.plot(x,y,x,y_est)
plt.show()

#preditor2
y_est = savepoint2.predict(x)
plt.figure(figsize=(16,8))
plt.plot(x,y,x,y_est2)
plt.show()

#preditor3
y_est = savepoint3.predict(x)
plt.figure(figsize=(16,8))
plt.plot(x,y,x,y_est3)
plt.show()
```

```python
lista = []
lista2 = []
lista3 = []

for lines in range(len(array1)):
    lista.append(array1[lines].loss_)

for lines in range(len(array2)):
    lista2.append(array2[lines].loss_)

for lines in range(len(array3)):
    lista3.append(array3[lines].loss_)
```

```python
numpyLista = np.asarray(lista)
numpyLista2 = np.asarray(lista2)
numpyLista3 = np.asarray(lista3)

print(numpyLista.mean())
print(numpyLista2.mean())
print(numpyLista3.mean())

print(numpyLista.std())
print(numpyLista2.std())
print(numpyLista3.std())
```
